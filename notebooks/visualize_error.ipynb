```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import pandas as pd
import pickle
import seaborn as sns
```

```python
sns.reset_orig()

plt.style.use('seaborn-paper')
mpl.rcParams.update({"axes.grid" : True})
```

# MMSys18 dataset

```python
res_dir = '../David_MMSys_18/DVMS/'
```

```python
plt.figure(figsize=(8, 6))
df = pd.DataFrame()
for d in sorted(os.listdir(res_dir)):
    if d[:8] == 'Results_':
        k = int(d.split('_')[-1])
        error_file = os.path.join(res_dir, d, 'errors.pkl')
        tmp = pd.DataFrame.from_records(pickle.load(open(error_file, 'rb')))
        tmp['K'] = [k] * len(tmp)
        df = df.append(tmp).reset_index(drop=True)
sns.lineplot(data=df, x='t', y='orthodromic_distance', hue='K', palette='flare_r')
plt.title('Test orthodromic distance')
plt.xlabel('prediction time step s (sec.)')
plt.ylabel('oethodromic distance (rad.)')
plt.legend(loc=(1.05, 0.0))
plt.show()
```

```python
for d in sorted(os.listdir(res_dir)):
    if d[:8] == 'Results_':
        k = int(d.split('_')[-1])
        error_file = os.path.join(res_dir, d, 'errors.pkl')
        df = pd.DataFrame.from_records(pickle.load(open(error_file, 'rb')))
        print(f'K={k}')
        ADE_str = ''
        for t in range(5):
            ADE_str += f"ADE {t + 1}s: {df[df['t'] <= t + 1]['orthodromic_distance'].mean():.3f} - "
        print(ADE_str[:-3])
```
